```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension ('tabulator')
import hvplot.pandas
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv')
```

```python
df.head()
```

```python
df
```

```python
df.columns
```

```python
df= df.fillna(0)
```

```python
df['GDP_per_capita']=np.where(df['population']!=0, df['gdp']/df['population'],0)
```

```python
idf=df.interactive()
```

```python
year_slider= pn.widgets.IntSlider(name='Year Slider', start= 1750, end=2020, step=5, value=1850)
```

```python
year_slider
```

```python
yaxis_co2=pn.widgets.RadioButtonGroup(
name= 'Y axis',
    options=['co2','co2_per_capita',],
    button_type='success'
)
```

```python
yaxis_co2
```

```python
continents= ['World','Asia','Oceania','Europe','Africa','North America', 'South America', 'Antarctica']
co2_pipeline=(
idf[
    (idf.year <= year_slider) &
    (idf.country.isin(continents))
]
.groupby(['country', 'year'])[yaxis_co2].mean()
.to_frame()
.reset_index()
.sort_values(by='year')
.reset_index(drop= True)
)
```

```python
co2_pipeline
```

```python
co2_plot= co2_pipeline.hvplot(x= 'year', by= 'country', y=yaxis_co2, line_width=2, title= "CO2 emission by continent")
```

```python
co2_plot
```

```python
co2_table= co2_pipeline.pipe(pn.widgets.Tabulator, pagination= 'remote', page_size=8, sizing_mode='stretch_width')
co2_table
```

```python
Co2_vs_GDP_scatterplot_pipeline= (
idf[(idf.year== year_slider) &
   (~(idf.country.isin(continents)))
   ]
    .groupby(['country', 'year', 'GDP_per_capita'])['co2'].mean()
  .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
  
)


# .to_frame()    .reset_index() .sort_values(by= 'year')  .reset_index(drop=True)
Co2_vs_GDP_scatterplot_pipeline 
```

```python
co2_vs_GDP_scatterplot= Co2_vs_GDP_scatterplot_pipeline.hvplot(x='GDP_per_capita', by='country', y= 'co2', kind= 'scatter', size= 80,legend= False, height= 500, width=500)
co2_vs_GDP_scatterplot
```

```python
co2_source_type=pn.widgets.RadioButtonGroup(
name= 'Y axis',
    options=['coal_co2','flaring_co2','oil_co2', 'gas_co2', 'cement_co2', 'other_industry_co2'],
    button_type='success'
)
co2_source_type
```

```python
continents_except_world= ['Asia','Oceania','Europe','Africa','North America', 'South America', 'Antarctica']
co2_source_pipeline=(
idf[
    (idf.year == year_slider) &
    (idf.country.isin(continents_except_world))
]
.groupby([ 'year', 'country'])[co2_source_type].sum()
.to_frame()
.reset_index()
.sort_values(by='year')
.reset_index(drop= True)
)
co2_source_pipeline
```

```python
co2_source_barplot= co2_source_pipeline.hvplot(x='country', y=co2_source_type,title= 'CO2 Source by Continent', kind= 'bar')
co2_source_barplot
```
